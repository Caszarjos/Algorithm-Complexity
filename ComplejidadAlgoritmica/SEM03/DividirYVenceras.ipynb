```python
import random

#SumarNormal
def sumarNumeros(arr):
    suma = 0
    for i in range(len(arr)):
        suma = suma + arr[i]
    return suma
#Recursivo Suma
def recursivoSuma(arr, n):
    if (n == 0): #caso base
        return arr[0]
    else:
        return arr[n] + recursivoSuma(arr, n-1)

#BuscarMayorNumero
def mayorNumero(arr):
    mayor = arr[0]
    for i in range(1,len(arr)): #[0,1,2,3,4,5,6,7.....]
        if(arr[i] > mayor):
            mayor = arr[i]
    return mayor
#MayorNumeroRecursivo
def recursivoMayor(arr, n, mayor): #(arr, 0, arr[0])
    if(n == len(arr)):
        return mayor
    else:
        if(arr[n] > mayor):
            mayor = arr[n]
        return recursivoMayor(arr, n+1, mayor)
    
#OtraFormaDeAplicarRecursividad
def max_recursivo(arr):
    if len(arr) == 1:  # caso base
        return arr[0]
    else:
        max_resto = max_recursivo(arr[1:])  # llamada recursiva
        return arr[0] if arr[0] > max_resto else max_resto
#RecursividadValorMinimo
def min_recursivo(arr):
    if len(arr) == 1:
        return arr[0]
    else:
        min_res = min_recursivo(arr[1:])
        return arr[0] if arr[0] < min_res else min_res

#Factorial
def factorial(num):
    fact = 1
    for i in range(1,num+1):
        fact = fact * i
    return fact
#FactorialRecursivo
def recursivoFactorial(num):
    if num == 0:
        return 1
    else:
        return num * recursivoFactorial(num - 1)

#Fibonacci n-esimo
#[1,1,2,3,5,8,13,21...]
def fibonacciRe(n):
    # Caso base
    if n == 0:
        return 0
    elif n == 1:
        return 1
    # Caso recursivo
    else:
        return fibonacci(n-1) + fibonacci(n-2)
def fibonacci(n):
    # Casos base
    if n == 0:
        return 0
    elif n == 1:
        return 1
    # Inicializar los primeros dos términos
    fib1, fib2 = 0, 1
    # Iterar para encontrar el n-ésimo término
    for i in range(2, n+1):
        fib = fib1 + fib2
        fib1, fib2 = fib2, fib
    return fib

#NumeroDeOcurrenciasEnUnaLista
def vecesRepetidas(arr, num):
    cont = 0
    for i in range(len(arr)):
        if(arr[i] == num):
            cont = cont +1
    return cont

def vecesRecursiva(arr, num):
    if len(arr) == 0:
        return 0
    else: 
        count = vecesRecursiva(arr[1:], num)
        if arr[0] == num:
            count += 1
        return count
#Palindromo
def es_palindromo(palabra):
    # Caso base: la palabra es vacía o de longitud 1
    if len(palabra) < 2: 
        return True
    # Caso recursivo: comparar el primer y último caracteres
    if palabra[0] == palabra[-1]: 
        # Llamada recursiva con la subpalabra que queda entre el primer y último caracteres
        return es_palindromo(palabra[1:-1])
    else:
        return False


def main():
    arr = []
    for i in range(200):
        arr.append(random.randint(1,100))
    #l = len(arr)
    #print(arr)
    #print(sumarNumeros(arr))
    #print(recursivoSuma(arr,l-1))
    #print(mayorNumero(arr))
    #print(recursivoMayor(arr, 0, arr[0]))
    #print(min_recursivo(arr))
    #print(recursivoFactorial(5))
    #print(fibonacci(4))
    #print(vecesRepetidas(arr,10))
    #print(vecesRecursiva(arr,10))
    print(es_palindromo("ANINA"))
main()
```

```python
def tot_palabras_por_linea(texto):
    if texto == "":
        return []
    lineas=texto.find("\n")
    if lineas == -1:
        return [len(texto.split())]
    
    lineaIzq = texto[:lineas]
    lineaDer = texto[lineas+1:]
    return [len(lineaIzq.split())] + tot_palabras_por_linea(lineaDer)
    
texto = "Lorem ipsum dolor sit amet, consectetur adipiscing elit.\n Nam arcu arcu, euismod mollis rhoncus quis, egestas at leo.\n Maecenas luctus neque nulla, vel imperdiet lacus ultrices vel."
print(tot_palabras_por_linea(texto))
```
