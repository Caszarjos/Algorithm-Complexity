Enumerar

```python
a = [4, 2, 33, 24, 54, 21, 199, 123]
for i, elem in enumerate(a):
    print(f"{i} -> {elem}")
```

Reversa

```python
for elem in reversed(a):
    print(elem)
```

Polinomio

```python
# x0: Variable del polinomio
# a: Lista de coeficientes del Polinomio
#def solve(x0, a):
def solve(x0, a :list):
    res = 0.0
    for i, ai in enumerate(reversed(a)):
        res += ai * x0 ** i		# ** es potencia, "i" es el grado de la variable

    return res

	
solve(2, [10, 20, 0, 15, 5, 50])
```
